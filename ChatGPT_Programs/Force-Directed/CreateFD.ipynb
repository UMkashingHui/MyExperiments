```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
edges = []
for i in range(80):
    source_nodes = random.sample(range(1, 80), 5)
    i_edges = []
    for node in source_nodes:
        if node == i:
            continue
        edge = tuple([i, node])
        i_edges.append(edge)
    for edge in i_edges:
        edges.append(edge)
```

```python
edges
```

```python
# Create an empty graph
options = {
    'node_color': 'blue',
    'node_size': 50,
    'width': 1,
}
G = nx.Graph()

# Define the edges in the graph
# edges = [(0, 1), (1, 2), (2, 3), (3, 4), (4, 0), (0, 3), (1, 4), (2, 4)]

# Add the edges to the graph
G.add_edges_from(edges)

# Compute the layout of the graph using the Fruchterman-Reingold force-directed algorithm
pos = nx.spring_layout(G)

# Draw the graph using the computed layout
nx.draw(G, pos, with_labels=False, **options)

# Show the graph
plt.savefig("fd.png")
```
