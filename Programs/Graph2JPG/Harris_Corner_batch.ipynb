```python
import numpy as np
import cv2 
import sys 
import time
import os
```

```python
#Read data file
def read_Graph(fname):
    data_list = [] #The data list
    num_nodes = int
    all_nodes = []
    num_edges = int
    all_edges = []
    node_x = []
    node_y = []
    data_f = open(fname, 'r')
    for line in data_f.readlines():
        data_list.append(line.strip())
    
    num_nodes = int (data_list.pop(0)) #Pop out the number of nodes 
    #Get all the number node coordinate
    for i in range (0,num_nodes):
        node = data_list.pop(0) #Pop out the node coordinate
        node_split = node.split()
        node_split.pop(0) #Pop out the node index  
        node_x.append(node_split[0])
        node_y.append(node_split[1])
    #Change it to list    
    node_x = list(map(float, node_x))
    node_y = list(map(float, node_y))
    #Find the minimun value in x and y list
    min_x = min(node_x)
    min_y = min(node_y)
    # TODO ??????????
    #If the coorodinate of nodes are nagative, change it to be positive
    if min_x or min_y < 0:
        for i in range (0,num_nodes):
            node_x[i] = node_x[i] + min_x * -1
            node_y[i] = node_y[i] + min_y * -1
    #Find the maximun value in x and y list        
    max_x = max(node_x)
    max_y = max(node_y)
    #Calulate the canvas size by the node numbers and radius of node
    canvas_size = (4 * num_nodes) # TODO ?????????????????
    #Calulate the ratio for enlarge the coorodinate of node that fitting the canvas size
    node_ratio = (min((canvas_size / max_x), (canvas_size / max_y)))
    #print(node_ratio)
    #Origin coordinates
    xc = 0
    yc = 0
    
    for i in range (0,num_nodes):
        x = int(xc + node_ratio * (node_x[i] - xc))
        y = int(yc + node_ratio * (node_y[i] - yc))
        all_nodes.append((x, y)) #(X, Y) 
    
    num_edges = int (data_list.pop(0)) #Pop out the number of edges 
    #Get all the edge
    for i in range (0,num_edges): 
        edge = data_list.pop(0) #Pop out the edge connection
        edge = edge.split()
        all_edges.append([int(edge[0]),int(edge[1])]) #(point 1, point 2)
    data_f.close
    # TODO ???????????????????????
    area_min_radio = 5000 / 4000
    area_min = area_min_radio * canvas_size
    
    return all_nodes, all_edges, num_nodes, canvas_size, area_min
```

```python
#Draw all the nodes
def drawAllNodes(img,all_nodes,radiu,color,thickness,lineType=cv2.LINE_AA,shift=0):
    for i in range (0,int(len(all_nodes))): 
        cv2.circle(img, all_nodes[i], radiu ,color,thickness,lineType,shift)  #Add the node result
        
    return img

#Draw all the edges
def drawAlledges(img,all_nodes,all_edges,color,thickness=3,lineType=cv2.LINE_AA,shift=0):
    for i in range (0,int(len(all_edges))):
        cv2.line(img, all_nodes[all_edges[i][0]], all_nodes[all_edges[i][1]],color,thickness, lineType, shift)
        
    return img
```

```python
def graph2bmp(graph_file_path, bmp_file_path):
    black = (0,0,0)
    white = (255,255,255)
    edge_color = black
    node_color = black
    canvas_color  = 255
    lineType= cv2.LINE_AA
    node_radius = 1
    
    all_nodes, all_edges, num_nodes, canvas_size, area_min = read_Graph(graph_file_path) 
    result_img = np.full((canvas_size, canvas_size, 3), canvas_color ,np.uint8) #create a img
    drawAlledges(result_img,all_nodes,all_edges,edge_color,lineType)
    drawAllNodes(result_img,all_nodes,node_radius,node_color,lineType);
    cv2.imwrite(bmp_file_path, result_img)
```

# Create folder in destination folder based on the original folder

```python
def mkdir(path):
    folder = os.path.exists(path)
    if not folder:
        os.makedirs(os.path.join(os.getcwd(), path))
        print("Create new folder "+os.path.join(os.getcwd(), path))
    else:
        print("Folder %s exists!"%os.path.join(os.getcwd(), path))
        
        
```

```python
def create_folder(ori_path, dst_path, ori_file_type, dst_file_type, dst_folder):
    ori_dirs = []
    dst_files_path = []
    ori_files_path = []
    for root, dirs, files in os.walk(ori_path, topdown=False):
        for name in dirs:
            dir_path = os.path.join(root,name)
            ori_dirs.append(dir_path.replace(ori_path, "") )
            
        for file in files:
            file_path = os.path.join(root, file)
            dst_file_path = file_path.replace(ori_path, os.path.join("data",dst_folder))
            dst_file_path = dst_file_path.replace(ori_file_type,dst_file_type)
            if ".DS_Store" not in dst_file_path:
                dst_files_path.append(dst_file_path)
                ori_files_path.append(file_path)

    for ori_dir in ori_dirs:
        mkdir(dst_path+'\\'+ori_dir)
        
    return ori_files_path,dst_files_path
```

```python
ori_path = os.path.join("data", "graph_files")
dst_path = os.path.join("data", "bmp_files")
graph_files_path, bmp_files_path = create_folder(ori_path, dst_path, ".graph", ".bmp", "bmp_files")
```

```python
def graph2bmp_batch():
    for graph, bmp in zip(graph_files_path, bmp_files_path):
        if os.path.exists(bmp):
            continue
        try:
            graph2bmp(graph, bmp)
        except:
            print(graph+" has some error!")
            break
    
```

```python
def bmp2jpg_batch():
    ori_path = os.path.join("data", "bmp_files")
    dst_path = os.path.join("data", "jpg_files")
    bmp_files_path, jpg_files_path = create_folder(ori_path, dst_path, ".bmp", ".jpg", "jpg_files")
    
    for bmp, jpg in zip(bmp_files_path, jpg_files_path):
        if os.path.exists(jpg):
            continue
        try:
            print(bmp)
            print(jpg)
            img = cv2.imread(bmp, -1)
            cv2.imwrite(jpg, img)
        except:
            print(bmp+" has some error!")
            break
    
```
